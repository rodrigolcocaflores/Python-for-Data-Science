# pandas groupby Demo

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../DATA/example_data.txt')
df
```

```python
df['mean_2_3'] = df[['value_2', 'value_3']].mean(axis=1)
df
```

```python
g = df.groupby(['run_id'])
for run_id, data in g:
    print(run_id)
    print(data)
    print()
```

```python
for run_id, data in g:
    print(data.describe(), '\n')
```

```python
for key, data in g:
    print('value_2 mean:', data['value_2'].mean())
    print('value_3 mean:', data['value_3'].mean())
    print('=' * 20)
```
